# Data Modeling

## Section 3 - Data Preparation

```python
def con_date(data: pd.DataFrame = None, column: str = None):
    df = data.copy()
    df[column] = pd.to_datetime(df[column], format = "%Y-%m-%d %H:%M:%S")
    return df
```

```python
sales_df = con_date(sales_df, 'timestamp')
```

```python
sales_df.info()
```

```python
stock_df = con_date(stock_df, 'timestamp')
```

```python
stock_df.info()
```

```python
temp_df = con_date(temp_df, 'timestamp')
```

```python
temp_df.info()
```

```python
def to_hourly(data: pd.DataFrame = None, column: str = None):
    df = data.copy()
    timeList = df[column].tolist()
    timeList = [i.strftime("%Y-%m-%d %H:00:00") for i in timeList ]
    timeList = [datetime.strptime(i, '%Y-%m-%d %H:00:00') for i in timeList]
    df[column] = timeList
    return df
```

```python
sales_df = to_hourly(sales_df, 'timestamp')
```

```python
sales_df.head()
```

```python
stock_df = to_hourly(stock_df, 'timestamp')
stock_df.head()
```

```python
temp_df = to_hourly(temp_df, 'timestamp')
temp_df.head()
```

```python
salesAgg = sales_df.groupby(["timestamp", "product_id"]).agg({'quantity' : 'sum'}).reset_index()
salesAgg.head()
```

```python
stockAgg = stock_df.groupby(["timestamp", "product_id"]).agg({"estimated_stock_pct" : "mean"}).reset_index()
stockAgg.head()
```

```python
tempAgg = temp_df.groupby(["timestamp"]).agg({'temperature' : 'mean'}).reset_index()
tempAgg.head()
```

```python
merged_df = stockAgg.merge(salesAgg, on=['timestamp', 'product_id'], how = 'left')
merged_df.head()
```

```python
merged_df = merged_df.merge(tempAgg, on = 'timestamp', how = 'left')
merged_df.head()
```

```python
merged_df.info()
```

```python
merged_df['quantity'].fillna(0, inplace = True)
merged_df.info()
```

```python
product_categories = sales_df[['product_id', 'category']]
product_categories = product_categories.drop_duplicates()

product_prices = sales_df[['product_id', 'unit_price']]
product_prices = product_prices.drop_duplicates()
```

```python
merged_df = merged_df.merge(product_categories, on="product_id", how = "left")
merged_df.head()
```

```python
merged_df = merged_df.merge(product_prices, on = "product_id", how = "left")
merged_df.head()
```

```python
merged_df.info()
```

## Section 4 - Feature Engineering

```python
merged_df['timestamp_day'] = merged_df['timestamp'].dt.day
merged_df['timestamp_week'] = merged_df['timestamp'].dt.dayofweek
merged_df["timestamp_hour"] = merged_df['timestamp'].dt.hour
merged_df.drop(columns="timestamp", inplace = True)
merged_df.head()
```

```python
merged_df = pd.get_dummies(merged_df, columns=['category'])
merged_df.head()
```

```python
merged_df.info()
```

Dropping `product_id` because it is not numerical and is simply an ID column

```python
merged_df.drop(columns="product_id", inplace = True)
merged_df.head()
```

## Section 5 - Modeling

```python
X = merged_df.drop(columns=['estimated_stock_pct'])
y = merged_df['estimated_stock_pct']
print(X.shape)
print(y.shape)
```

```python
K = 10
split = 0.8
```

Random Forest Regressor for modeling a continous numeric variable.

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
from sklearn.preprocessing import StandardScaler
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = split, random_state = 42)
```

```python
# Model
model = RandomForestRegressor()
scaler = StandardScaler()

# Scaling X data
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# Training the model
trained_model = model.fit(X_train, y_train)

# Model Prediction
y_pred = trained_model.predict(X_test)
```

```python
# computing model accuracy
mae = mean_absolute_error(y_test, y_pred)
print("Accuracy: ", mae)
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
features = [i.split("__")[0] for i in X.columns]
importances = model.feature_importances_
indices = np.argsort(importances)
```

```python
fig, ax = plt.subplots(figsize=(20,20))
plt.title("Feature importance")
plt.barh(range(len(indices)), importances[indices], color='b', align='center' )
plt.yticks( range(len(indices)), [features[i] for i in indices] )
plt.xlabel("Relative importance")
plt.show()
```
